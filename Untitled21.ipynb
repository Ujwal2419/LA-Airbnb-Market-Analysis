```python
df=df.drop(columns='price')
```

```python
# Fill null reviews with 0 (they're new listings)
df['last_review'] = df['last_review'].fillna('Never')
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

```python
# Fill with placeholder
df['host_name'] = df['host_name'].fillna('Unknown Host')
```

```python
# Create a binary flag instead
df['has_license'] = df['license'].notna().astype(int)
```
